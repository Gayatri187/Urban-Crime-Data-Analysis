#Crime density heatmap


```python
from folium.plugins import HeatMap
heat_data = df[['Latitude','Longitude']].dropna().values.tolist()
crime_map = folium.Map(location=[41.8781,-87.6298],zoom_start=11)
HeatMap(heat_data[:2000]).add_to(crime_map)
crime_map
```

What this does

takes only latitude and longitude

removes missing values

creates a fresh Chicago map

adds a heatmap using the first 2000 valid points

What you should see

blue/green/yellow/red density areas

brighter areas = more crime concentration

#Crime trend by month

```python
df['Date']=pd.to_datetime(df['Date'])
df['Month']=df['Date'].dt.month
df['Month'].value_counts().sort_index()
```

```python
df['Month'].value_counts().sort_index().plot(kind='bar')
plt.title("Crime distribution per Month")
plt.xlabel("Month")
plt.ylabel("Number of crimes ")
plt.show()
```

#Crime by district

```python
df['District'].value_counts().head(10)
```

```python
df['District'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Districts with Highest Crime")
plt.xlabel("District")
plt.ylabel("Crime Count")
plt.tight_layout()
plt.savefig("top_10_districts_with_highcrime.png")
plt.xticks(rotation=0)
plt.show()
```

#Crime by hour

```python
df['Hour'] = df['Date'].dt.hour

df['Hour'].value_counts().sort_index()
```

```python
df['Hour'].value_counts().sort_index().plot(kind='line', marker='o')

plt.title("Crime Distribution by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Crimes")
plt.tight_layout()
plt.savefig("crime_distribution_by_hour.png")
plt.show()
```

#Key Insights
1. Theft and battery are among the most common crime types
